```python
import json
import pickle

# JSON íŒŒì¼ ë¡œë“œ
json_path = "../ë°ì´í„°/highschool_english_dataset.json"
with open(json_path, "r", encoding="utf-8") as f:
    dataset = json.load(f)

# ì§ˆë¬¸ê³¼ ì •ë‹µì„ ë¦¬ìŠ¤íŠ¸ë¡œ ë³€í™˜
faiss_data = []
for _, content in dataset.items():
    questions = content.get("questions", {}).get("cleaned_text", "").split("\n")
    answers = content.get("answers", {})

    for i, question in enumerate(questions, 1):
        if str(i) in answers:  # ì •ë‹µì´ ì¡´ì¬í•˜ëŠ” ê²½ìš°
            faiss_data.append({"question": question.strip(), "answer": answers[str(i)]})

# `faiss_data.pkl` ì €ì¥
pkl_path = "faiss_data.pkl"
with open(pkl_path, "wb") as f:
    pickle.dump(faiss_data, f)

print(f"âœ… {pkl_path} ìƒì„± ì™„ë£Œ! ì´ {len(faiss_data)}ê°œì˜ ë°ì´í„°ê°€ ì €ì¥ë¨.")
```

```python
import json
import pickle
import re

# âœ… JSON íŒŒì¼ ê²½ë¡œ
json_path = "../ë°ì´í„°/highschool_english_dataset.json"

# âœ… JSON íŒŒì¼ ë¡œë“œ
with open(json_path, "r", encoding="utf-8") as f:
    json_data = json.load(f)

# âœ… ëª¨ë“  ë¬¸ì œ ë°ì´í„°ë¥¼ ë¦¬ìŠ¤íŠ¸ë¡œ ë³€í™˜
faiss_data = []

# ğŸ” JSONì˜ ê° íŒŒì¼(ì‹œí—˜ì§€) ìˆœíšŒ
for key, content in json_data.items():
    if isinstance(content, dict) and "questions" in content and "answers" in content:
        raw_text = content["questions"].get("raw_text", "")
        answers = content["answers"]
        
        # âœ… ë¬¸ì œ ë²ˆí˜¸ + ì„ íƒì§€ ì¶”ì¶œ (ì •ê·œ í‘œí˜„ì‹ ì‚¬ìš©)
        question_pattern = r"(\d+)\.\s(.+?)(?=\s\d+\.\s|$)"  # ë¬¸ì œ ë²ˆí˜¸ë¥¼ ê¸°ì¤€ìœ¼ë¡œ ë¶„ë¦¬
        matches = re.findall(question_pattern, raw_text, re.DOTALL)

        for q_number, q_text in matches:
            q_number = q_number.strip()  # ë¬¸ì œ ë²ˆí˜¸ ì •ë¦¬
            q_text = q_text.strip()  # ë¬¸ì œ ë‚´ìš© ì •ë¦¬
            
            # âœ… ì„ íƒì§€ ì¶”ì¶œ (â‘  ~ â‘£ í˜•íƒœ)
            option_pattern = r"â‘  (.*?) â‘¡ (.*?) â‘¢ (.*?) â‘£ (.*?)$"
            option_match = re.search(option_pattern, q_text, re.DOTALL)

            if option_match:
                options = list(option_match.groups())  # ì„ íƒì§€ ë¦¬ìŠ¤íŠ¸ë¡œ ë³€í™˜
                q_text = re.sub(option_pattern, "", q_text).strip()  # ì„ íƒì§€ ë¶€ë¶„ ì œê±°í•œ ë¬¸ì œ í…ìŠ¤íŠ¸
            else:
                options = []  # ì„ íƒì§€ê°€ ì—†ì„ ê²½ìš° ë¹ˆ ë¦¬ìŠ¤íŠ¸

            # âœ… ì •ë‹µ ë§¤í•‘
            correct_answer = answers.get(q_number, None)
            if correct_answer:
                try:
                    correct_answer = int(correct_answer) - 1  # 1~4 -> 0~3 ì¸ë±ìŠ¤ë¡œ ë³€í™˜
                    correct_answer = options[correct_answer] if 0 <= correct_answer < len(options) else None
                except ValueError:
                    correct_answer = None

            # âœ… ë°ì´í„° ì €ì¥
            if options and correct_answer:
                faiss_data.append({
                    "question": q_text,
                    "options": options,
                    "answer": correct_answer
                })

# âœ… `faiss_data.pkl` ì €ì¥
faiss_data_path = "faiss_data.pkl"
with open(faiss_data_path, "wb") as f:
    pickle.dump(faiss_data, f)

print(f"âœ… `faiss_data.pkl` ìƒì„± ì™„ë£Œ! ì´ ë¬¸ì œ ê°œìˆ˜: {len(faiss_data)}")
```
